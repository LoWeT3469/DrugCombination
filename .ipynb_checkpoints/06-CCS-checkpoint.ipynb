### Cell Line - Cell Line Similarity

For two distinct cell lines a and b where $a \neq b$, the cell line similarity $s_{a,b} = d^{âˆ’1}_{a,b}$ was calculated as the inverse of Euclidean distance $d_{a,b}$ between the two vectors $V_{a}$ and $V_{b}$ from $\color{yellow}{\text{gene expression, microRNA expression, and protein abundance}}$, and the cell line similarity matrix $(M_{Y Y} )$ was created, where the off-diagonal $(a \neq b)$ element $e_{a,b} = e_{b,a}$ is the cell line similarity value $s_{a,b} = s_{b,a}$, min-max normalized to values between 0 and 1, inclusive. The diagonal elements $e_{a,a}$ were set to 1.

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import euclidean

# Read the CSV file containing the data
df = pd.read_csv("60CellLinesDescriptor.csv")

# Select only the last 60 columns as features
features = df.iloc[:, -60:].values

# Calculate Euclidean distance between each pair of vectors
distances = np.zeros((len(features), len(features)))
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        distances[i, j] = euclidean(features[i], features[j])
        distances[j, i] = distances[i, j]  # Symmetric matrix

# Set infinite values and NaNs to 0
distances[np.isinf(distances) | np.isnan(distances)] = 0

# Compute the inverse of the Euclidean distance
inverse_distances = np.reciprocal(distances, where=distances != 0)

# Normalize the similarity values between 0 and 1
max_inverse_distance = np.nanmax(inverse_distances)
inverse_distances_normalized = inverse_distances / max_inverse_distance

# Set diagonal elements to 1
np.fill_diagonal(inverse_distances_normalized, 1)

# Create a DataFrame for the similarity matrix
similarity_matrix = pd.DataFrame(inverse_distances_normalized, index=df.index, columns=df.index)

# Display the similarity matrix
print(similarity_matrix)
```
